```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("../data/raw_dataset.csv")

df.head()
```

```python
print("Dataset Shape:", df.shape)
print("\nColumn Info:")
print(df.info())
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
df["impact"].value_counts()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="impact", order=["low","medium","high"])
plt.title("Impact Class Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="transport", hue="impact")
plt.title("Transport vs Impact")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="electricity", hue="impact")
plt.title("Electricity Usage vs Impact")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="recycling", hue="impact")
plt.title("Recycling vs Impact")
plt.show()
```

```python
encoded_df = df.copy()

for col in ["transport", "electricity", "recycling", "meat_consumption", "water_usage", "impact"]:
    encoded_df[col] = encoded_df[col].astype("category").cat.codes

encoded_df.corr()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(encoded_df.corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```
